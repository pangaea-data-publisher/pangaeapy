Its a series so lets take a look how many PANGAEA 'events' (= series identifier) this data set has...

```python
for ev in ds.events:
    print(ev.label)
```

We set an index on the 'Event' column to be able to use the Event name as a selector later on..

```python
ds.data.set_index('Event',inplace=True)
```

Now we pick one of the events to plot a temperature, salinity profile

```python
ds.data.loc['PS04/223-1'].plot(x='Depth water', y=['Temp','Sal'],style='-', secondary_y='Sal')
ds.data.loc['PS04/223-1'].plot(x='Depth water', y=['Sal'],style='.-')
```
